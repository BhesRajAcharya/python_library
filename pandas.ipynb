```python
#merge function of pandas
```

```python
import pandas as pd
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4,5],'class':[10,11,12,13,14]})
df1
```

```python
df2=pd.DataFrame({'ID':[2,4,6,8,10],'name':['bijaya','udaya','arjun','ramesh','sagar']})
df2
```

```python
pd.merge(df1,df2,on='ID')
```

```python
pd.merge(df1,df2,how='left')
```

```python
pd.merge(df1,df2,how='right')
```

```python
pd.merge(df1,df2,how='outer',indicator=True)
```

```python
df2=pd.DataFrame({'ID':[6,7,8,9,10],'name':['bijaya','udaya','arjun','ramesh','sagar']})
df2
```

```python
pd.merge(df1,df2,left_index=True,right_index=True)
```
